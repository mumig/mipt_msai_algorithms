```python
def bubble_sort(nums):
    # Устанавливаем swapped в True, чтобы цикл запустился хотя бы один раз
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(nums) - 1):
            if nums[i] > nums[i + 1]:
                # Меняем элементы
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                # Устанавливаем swapped в True для следующей итерации
                swapped = True


n = int(input())
stud_lst = []

for i in range(n):
    stud_id, score, name = input().split()
    stud_id = int(stud_id)
    score = -int(score)
    stud_lst.append((score, stud_id, name))

bubble_sort(stud_lst)

for i in range(3):
    print(stud_lst[i][2])

print(*[stud_lst[i][1] for i in range(len(stud_lst))])
```

```python
def partition(nums, low, high):

    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        nums[i], nums[j] = nums[j], nums[i]


def quick_sort(nums):

    def _quick_sort(items, low, high):
        if low < high:
            split_index = partition(items, low, high)
            _quick_sort(items, low, split_index)
            _quick_sort(items, split_index + 1, high)

    _quick_sort(nums, 0, len(nums) - 1)


n = int(input())
lst = list(map(int, input().split()))
quick_sort(lst)
print(*lst)
```

```python
def partition(nums, low, high):

    pivot = nums[(low + high) // 2]
    i = low - 1
    j = high + 1
    while True:
        i += 1
        while nums[i] < pivot:
            i += 1

        j -= 1
        while nums[j] > pivot:
            j -= 1

        if i >= j:
            return j

        nums[i], nums[j] = nums[j], nums[i]


def quick_sort(nums):

    def _quick_sort(items, low, high):
        if low < high:
            split_index = partition(items, low, high)
            _quick_sort(items, low, split_index)
            _quick_sort(items, split_index + 1, high)

    _quick_sort(nums, 0, len(nums) - 1)


n, g = map(int, input().split())

res = 0
eff_lst = []
for i in range(n):
    c, w = map(int, input().split())
    if w != 0:
        eff_lst.append((c / w, w, c))
    else:
        res += c

quick_sort(eff_lst)

empty_g = g
i = 1
while empty_g > 0 and i <= len(eff_lst):
    w = min(empty_g, eff_lst[-i][1])
    res += eff_lst[-i][2] * w // eff_lst[-i][1]
    empty_g -= w
    i += 1
print(res)
```
