## Regular Bracket Sequence

image.png

image.png

```python
def rbin_search(lst, x):

    if x >= lst[-1]:
        return len(lst)

    left_bor = 0
    right_bor = len(lst)
    while left_bor != right_bor:
        middle = int((left_bor + right_bor) / 2)
        if lst[middle] <= x:
            left_bor = middle + 1
        else:
            right_bor = middle
        # print(left_bor, right_bor, middle)
    return left_bor


def lbin_search(lst, x):

    if x > lst[-1]:
        return len(lst)

    left_bor = 0
    right_bor = len(lst)
    while left_bor != right_bor:
        middle = int((left_bor + right_bor) / 2)
        if x <= lst[middle]:
            right_bor = middle
        else:
            left_bor = middle + 1
        # print(left_bor, right_bor, middle)
    return left_bor


n, m = map(int, input().split())
lst = sorted(list(map(int, input().split())))
for i in range(m):
    l, r = map(int, input().split())
    if len(lst) == 0:
        print(0)
    else:
        print(rbin_search(lst, r) - lbin_search(lst, l))
```

```python
inp = list(map(int, input().split()))
request = inp[0]

queue = []
while request != -1:
    if request == 0:
        queue.append(inp[-1])
    elif request == 1:
        if len(queue) == 0:
            queue.append(inp[-1])
        else:
            queue.insert(0, inp[-1])
    elif request == 2:
        print(len(queue))
    elif request == 3:
        if len(queue) == 0:
            print('Error!')
        else:
            print(queue[-1])
            queue.pop(-1)
    elif request == 4:
        if len(queue) == 0:
            print('Error!')
        else:
            print(queue[0])
            queue = queue[1:]

    inp = list(map(int, input().split()))
    request = inp[0]
```

```python
n = int(input())

queues = []
for i in range(n):
    queues.append([])

inp = list(input().split())
request = inp[0]

while request != '#':
    shop_ind = int(inp[1])
    if request == '+':
        queues[shop_ind].append(int(inp[-1]))
    elif request == '-':
        print(queues[shop_ind][0])
        queues[shop_ind].pop(0)
    elif request == '!':
        ln = len(queues[shop_ind])
        middle = int(ln / 2) + ln % 2
        queues[shop_ind].insert(middle, int(inp[-1]))
    elif request == '?':
        print(len(queues[shop_ind]))

    inp = list(input().split())
    request = inp[0]
```
