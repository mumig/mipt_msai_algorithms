```python
def UpperBound(A, key):
    left = -1
    right = len(A)
    while right > left + 1:
        middle = (left + right) // 2
        if A[middle] >= key:
            right = middle
        else:
            left = middle
    return right


n = int(input())
if n < 2:
    print(n)
else:
    a = list(map(int, input().split()))
    d = [1000000] * (n + 1)
    d[0] = -1
    for i in range(n):
        ind = UpperBound(d, a[i])
        d[ind] = a[i]
    i = 0
    while d[i] != 1000000 and i != n:
        i += 1
    if i == n:
        print(i)
    else:
        print(i - 1)
```

```python
n = int(input())
a = list(map(int, input().split()))
a = sorted(a)
if n == 2:
    print(a[1] - a[0])
else:
    d = [0] * (n - 1)
    d[0] = a[1] - a[0]
    d[1] = a[2] - a[0]
    for i in range(3, n):
        d[i - 1] = min(d[i - 2], d[i - 3]) + a[i] - a[i - 1]
    print(d[-1])
```
