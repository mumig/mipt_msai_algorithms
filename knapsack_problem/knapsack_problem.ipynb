```python
def ksp(W, N, weights, costs):

    K = [[0 for x in range(W + 1)] for x in range(N + 1)]
    cur_inds = ['' for x in range(W + 1)]
    ln = 0

    for i in range(1, N + 1):
        last_inds = cur_inds
        cur_inds = ['' for x in range(W + 1)]
        for j in range(1, W + 1):
            if weights[i - 1] <= j:
                add_cur = costs[i - 1] + K[i - 1][j - weights[i - 1]]
                do_nothing = K[i - 1][j]
                if add_cur >= do_nothing:
                    cur_inds[j] = last_inds[j - weights[i - 1]] + str(i) + ' '
                    K[i][j] = add_cur
                else:
                    cur_inds[j] = last_inds[j]
                    K[i][j] = do_nothing
            else:
                K[i][j] = K[i - 1][j]
                cur_inds[j] = last_inds[j]

    return K[-1][-1], list(cur_inds[-1].split())


W, N = map(int, input().split())

weights = list(map(int, input().split()))
costs = list(map(int, input().split()))

costs, inds = ksp(W, N, weights, costs)

print(costs)

if len(inds) == 0:
    print(0)
else:
    print(len(inds))
    print(*inds)
```

```python
def is_dev(lst):

    cost = sum(lst)
    if cost % 2 != 0:
        return 'NO'
    else:
        cost = cost // 2

    K = [[0 for x in range(cost + 1)] for x in range(len(lst) + 1)]
    for i in range(len(lst) + 1):
        for j in range(cost + 1):
            if i == 0 or j == 0:
                K[i][j] = 0
            elif lst[i - 1] <= j:
                K[i][j] = max(lst[i - 1] + K[i - 1][j - lst[i - 1]], K[i - 1][j])
            else:
                K[i][j] = K[i - 1][j]

    if K[-1][-1] == cost:
        res = 'YES'
    else:
        res = 'NO'
    return res


n = int(input())
lst = list(map(int, input().split()))
print(is_dev(lst))
```
