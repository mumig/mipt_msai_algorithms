## Distinct substrings

```python
def c_to_i(ch):
    return ord(ch) - ord('a') + 1


def substr(s, a=321, p=59604644783353249):
    N = len(s)
    a_pow = [1] * (N + 1)
    for i in range(N):
        a_pow[i + 1] = (a_pow[i] * a) % p
    hashes = set()
    for i in range(N):
        h = 0
        for j in range(i, N):
            h = (h + c_to_i(s[j]) * a_pow[j - i]) % p
            hashes.add(h)
    return len(hashes)


s = input()
print(substr(s))
```
