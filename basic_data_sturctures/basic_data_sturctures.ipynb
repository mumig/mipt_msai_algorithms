## Regular Bracket Sequence

```python
string = str(input())

open_symbols = set(['(', '[', '{'])
match_symb = {'(': ')', '[': ']', '{': '}', 'empty': 'empty'}

result = 'yes'
opened = ['empty']

for symb in string:
    if symb in open_symbols:
        opened.append(symb)
    elif symb == match_symb[opened[-1]]:
        opened.pop(-1)
    else:
        result = 'no'
        break

if len(opened) != 1:
    result = 'no'

print(result)
```

```python
inp = list(map(int, input().split()))
request = inp[0]

queue = []
while request != -1:
    if request == 0:
        queue.append(inp[-1])
    elif request == 1:
        if len(queue) == 0:
            queue.append(inp[-1])
        else:
            queue.insert(0, inp[-1])
    elif request == 2:
        print(len(queue))
    elif request == 3:
        if len(queue) == 0:
            print('Error!')
        else:
            print(queue[-1])
            queue.pop(-1)
    elif request == 4:
        if len(queue) == 0:
            print('Error!')
        else:
            print(queue[0])
            queue = queue[1:]

    inp = list(map(int, input().split()))
    request = inp[0]
```

```python
n = int(input())

queues = []
for i in range(n):
    queues.append([])

inp = list(input().split())
request = inp[0]

while request != '#':
    shop_ind = int(inp[1])
    if request == '+':
        queues[shop_ind].append(int(inp[-1]))
    elif request == '-':
        print(queues[shop_ind][0])
        queues[shop_ind].pop(0)
    elif request == '!':
        ln = len(queues[shop_ind])
        middle = int(ln / 2) + ln % 2
        queues[shop_ind].insert(middle, int(inp[-1]))
    elif request == '?':
        print(len(queues[shop_ind]))

    inp = list(input().split())
    request = inp[0]
```
